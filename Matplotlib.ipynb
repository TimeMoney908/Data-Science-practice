```python
fig,ax = plt.subplots(figsize=(12,4))

plt.plot(df["Date"], df["GSPC500"])
plt.xlabel("Date")
plt.ylabel("Stock Price")
```

```python
fig,ax = plt.subplots(figsize=(12,4))

plt.plot(df["Date"], df["GSPC500"])
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("Stock Trend")
```

```python
fig,ax = plt.subplots(figsize=(10,5))

plt.plot(df["Date"], df["GSPC500"])
plt.plot(df["Date"], df["FTSE100"])
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("Stock Trend")
```

```python
label = ["GSPC", "FTSE"]
```

```python
fig,ax = plt.subplots(figsize=(12,4))

plt.plot(df["Date"], df["GSPC500"])
plt.plot(df["Date"], df["FTSE100"])
plt.xlabel("Date")
plt.ylabel("Stock Price")
plt.title("Stock Trend")
plt.legend(labels = label, fontsize="large")
```

## Bar chart

```python
df = pd.read_csv("bar_chart_data.csv")
```

```python
df.head()
```

```python
fig,ax = plt.subplots(figsize=(12,4))
plt.bar(df["Brand"], df["Cars Listings"])
```

```python
fig,ax = plt.subplots(figsize=(12,4))
plt.bar(df["Brand"], df["Cars Listings"], color = "darkorange")
plt.xlabel("Car names")
```

```python
fig,ax = plt.subplots(figsize=(9,4))
plt.bar(df["Brand"], df["Cars Listings"], color = "darkorange")
plt.xlabel("Car names")
plt.xticks(rotation = 45)
```
