### Sqlite3

```python
## connect to a database

import sqlite3
```

```python
db = sqlite3.connect("flights.db")
```

```python
# create a cursor object

cur = db.cursor()
```

```python
cur.execute("create table dailyflighte (id integer, name varchar(10), passn varchar(10))")
db.commit()
```

```python
## once table is created let me insert data

cur.execute("insert into dailyflighte values(1, 'xy', 'yt')")
db.commit()
```

```python
df1 = pd.read_sql_query("Select * from dailyflighte;", db)
```
