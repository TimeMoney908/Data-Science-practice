# Line plot

```python
df= pd.DataFrame(np.random.randn(100,3), columns=['A','B', 'C'])
df.head()
```

```python
df = df.cumsum()
```

```python
df.head()
```

```python
df.shape
```

```python
df.iplot()
```

```python
df.plot()
```

# scatter plot 

```python
df.head()
```

```python
df.iplot(x = 'A', y ='B', mode = 'markers', size = 25, title = 'Scatter Plot', xTitle ='X', yTitle = 'Y')
```

# Bar plot 

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
titanic.iplot(kind = 'bar', x = 'sex', y = 'survived', title = 'Sexy pooja')
```

```python
titanic['sex'].value_counts()
```

```python
df.iplot(kind = 'bar')
```

```python
df.iplot(kind = 'bar', barmode = 'stack',bargap = 0.5)
```

```python
df.iplot(kind = 'barh', barmode = 'stack',bargap = 0.5)
```

# Box plot & area plot
